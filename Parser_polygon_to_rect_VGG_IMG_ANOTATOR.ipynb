```python
from jsonmerge import merge
import functools
lista=[fant]#dvera,fgon,fant,rsten]
c=functools.reduce(lambda x,y: merge(x,y),lista) #parsing a list of list of dictionary in a big one dictionary
b=c.copy()
```

```python
for key, value in c.items():
    for k,v in value["regions"].items():
        if(v["shape_attributes"]["name"]=="polygon"):
            
            x_min=min(c[key]["regions"][k]["shape_attributes"]["all_points_x"])
            x_max=max(c[key]["regions"][k]["shape_attributes"]["all_points_x"])
            y_min=min(c[key]["regions"][k]["shape_attributes"]["all_points_y"])
            y_max=max(c[key]["regions"][k]["shape_attributes"]["all_points_y"])

            
            c[key]["regions"][k]["shape_attributes"]["x"]=x_min
            c[key]["regions"][k]["shape_attributes"]["y"]=y_min
            c[key]["regions"][k]["shape_attributes"]["width"]=x_max-x_min
            c[key]["regions"][k]["shape_attributes"]["height"]=y_max-y_min
            
            c[key]["regions"][k]["shape_attributes"].pop("all_points_x")
            c[key]["regions"][k]["shape_attributes"].pop("all_points_y")
            
            c[key]["regions"][k]["shape_attributes"]["name"]="rect"
            

            
```
